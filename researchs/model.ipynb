````python
# Loading Data
df = pd.read_csv('datasets/WineQT.csv')
````

````python
# Id is Not required so drop it 
df = df.drop('Id', axis=1)
````

````python
df.head()
````

````python
df['quality'].unique()
````

````python
df['quality'].mean()
````

##### df['quality'] Contains array([5, 6, 7, 4, 8, 3]) of items
now we can change this into 0 and 1
```<=5 is 0 --> means Bad Quality```
```>5 is 1 --> means Good Quality```

````python
df['label'] = df['quality'].apply(lambda z: 0 if z<=5 else 1)
````

````python
df.head()
````

````python
df['label'].unique()
````

### Spliting Data

````python
x = df.drop(['quality', 'label'], axis=1)# --> Features
y = df['label']# --> Label or Target 
````

````python
x.head()
````

````python
y
````

````python
x.shape, y.shape # X --> 2d  Y --> 1D
````

We want split data for training and testing

````python
x_train,x_test,y_train,y_test = train_test_split(
    x,y,
    test_size=0.2,# 20% data for testing
    stratify=y,
    random_state=42
)
````

````python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
````

## Model Building

````python
model = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    min_samples_split=2,
    max_features='sqrt',
    bootstrap=True,
    n_jobs=-1, 
    random_state=42
)
````

````python
model.fit(x_train,y_train)
````

## Model Score

````python
y_pred = model.predict(x_test)
y_pred
````

````python
score_acc = accuracy_score(y_test, y_pred)
score_clas = classification_report(y_test, y_pred)
print(f'Accuracy : {score_acc}')
print(score_clas)
````

## Importance of Features

````python
feat_imp = pd.Series(
    model.feature_importances_,
    index=x.columns
).sort_values(ascending=False)
feat_imp
````
